```python
# libraires needed

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import mglearn
from sklearn.preprocessing import OneHotEncoder
```

Using the adult dataset from mglearn chapter 4 

```python
adult_path = os.path.join(mglearn.datasets.DATA_PATH, "adult.data")


data = pd.read_csv(
    adult_path, header=None, index_col=False,
    names=['age', 'workclass', 'fnlwgt', 'education',  'education-num',
           'marital-status', 'occupation', 'relationship', 'race', 'gender',
           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
           'income'])

data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week',
             'occupation', 'income']]

data.head()
```

```python

# Identifying categorical and numerical variables.

# Created lists to put the values that are categorical and numerical
categorical = []
numerical = []


# Created a for loop to look get the correct values into the list
for i in data:
    data_type = data[i].dtype
    if data_type == "str":
        categorical.append(i)
    elif data_type == int:
        numerical.append(i)
    

print(f"Categorical features in this data set include: \n{categorical}")
print('')
print(f"Numerical features in this data set include: \n{numerical}")
```

```python
# visualizing histogram for age 
```

```python


education_VC = data["education"].value_counts()
counts = education_VC.index
values = education_VC.values

plt.bar(counts, values)
plt.xticks(rotation=65) 
plt.show
```

```python
# data = data[[ 'workclass', 'education', 'gender','occupation', 'income']]
data_dummies = pd.get_dummies(data)
# encoded data with 0 or 1 instead of bool


features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']

X = features.values
y = data_dummies["income_ >50K"].values
# data_dummies
print(f"shape of data \n X.shape:{X.shape}, y.shape:{y.shape}")

data_dummies.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print("Test score: {:.2f}".format(logreg.score(X_test, y_test)))
```

```python
# seeing how age and income relate


data['age'] = pd.cut(
    data['age'],
    bins=[0, 25, 35, 45, 55, 100],
    labels=['<25', '2534', '3544', '4554', '55+'])
```

```python
data = pd.get_dummies(data, columns=['age'], drop_first=True)
```

```python

# getting income data as X, y
X = data.drop('income', axis=1)
y = data['income']

# clean strings
for col in X.select_dtypes(include='object').columns:
    X[col] = X[col].str.strip()

#  One hot encoding
X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
score = logreg.score(X_test, y_test)
print(f"model has {score:.2f}% accuracy on unseen data")
```
